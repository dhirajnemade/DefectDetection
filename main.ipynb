```python
import os
import numpy as np
from MVTecADLoader import MVTecADLoader
from CNNModel import CNNModel
from CNN_RNN_Model import CNN_RNN_Model
from GAN import GAN
from GANForDefectDetection import GANForDefectDetection
from defect_detection import DefectDetection

def main(product,data_dir, model_type='cnn', epochs=10, batch_size=32, sequence_length=5, latent_dim=100, save_interval=200):
    # Load the dataset for CNN or CNN-RNN
    if model_type in ['cnn', 'cnn_rnn']:
        data_loader = MVTecADLoader(data_dir)
        train_images, train_labels, val_images, val_labels = data_loader.get_data()
        class_names = os.listdir(data_dir)  # Assuming class names match directory names

        if model_type == 'cnn':
            # Define the CNN model
            input_shape = train_images.shape[1:]  # (128, 128, 3) for example
            num_classes = train_labels.shape[1]  # Number of classes
            model = CNNModel(input_shape, num_classes)

            # Train the model
            history = model.train(train_images, train_labels, val_images, val_labels, epochs=epochs, batch_size=batch_size)

            # Plot training history
            model.plot_training_history(history)

            # Evaluate the model with a confusion matrix and classification report
            model.plot_confusion_matrix_and_report(val_images, val_labels, class_names)
            

        elif model_type == 'cnn_rnn':
            # Define the CNN-RNN model
            cnn_rnn_model = CNN_RNN_Model(data_dir, image_size=(128, 128), sequence_length=sequence_length)
            cnn_rnn_model.load_data()

            # Create the CNN-RNN model architecture
            cnn_rnn_model.create_cnn_rnn_model()

            # Train the model
            history = cnn_rnn_model.train_model(epochs=epochs, batch_size=batch_size)

            # Evaluate the model
            cnn_rnn_model.evaluate_model()

            # Plot the training history
            cnn_rnn_model.plot_training_history(history)

            # Generate and plot the confusion matrix and classification report
            cnn_rnn_model.plot_confusion_matrix_and_report(class_names)
            
    elif model_type == 'dd':
            # Use the DefectDetection class to handle CNN-based defect detection
            good_images_dir = os.path.join(data_dir, product+'/train/')
            defected_images_dir = os.path.join(data_dir, product+'/test/')
            test_images_dir = os.path.join(data_dir, product+'/test/')
            model_save_path = 'defect_detection_model.h5'

            classifier = DefectDetection()

            # Step 1: Train the model on both good and defected images
            print("Training the CNN model on good and defected images...")
            classifier.train_on_images(good_images_dir, defected_images_dir, model_save_path, epochs=epochs, batch_size=batch_size)

            # Step 2: Test the model on test images
            print("Testing the model on test images...")
            classifier.classify_test_images(test_images_dir, model_save_path)

    elif model_type == 'gan':
        # Example: Create dummy training images (replace with actual training images)
        # For GAN, we'll assume that train_images has already been loaded
        train_images = np.random.rand(1000, 128, 128, 3)  # Replace with your actual training data

        # Instantiate the GAN class
        img_shape = (128, 128, 3)
        gan_model = GAN(latent_dim=latent_dim, img_shape=img_shape)

        # Print the models' summaries
        gan_model.generator.summary()
        gan_model.discriminator.summary()
        gan_model.gan.summary()

        # Train the GAN
        gan_model.train(train_images=train_images, epochs=epochs, batch_size=batch_size)

    elif model_type == 'gan_for_defect_detection':
        # Train GAN for defect detection
        gan_for_defect_detection = GANForDefectDetection(data_dir, image_size=(64, 64), noise_dim=latent_dim)
        gan_for_defect_detection.train(epochs=epochs, batch_size=batch_size, save_interval=save_interval)
        gan_for_defect_detection.evaluate()

    else:
        print(f"Unsupported model type: {model_type}. Please choose 'cnn', 'cnn_rnn', 'gan', or 'gan_for_defect_detection'.")

if __name__ == "__main__":
    data_dir = 'dataset/1234/Oring'  # Replace with your actual dataset path
    product='Oring'
    model_type = 'cnn'  # Choose between 'cnn','dd', 'cnn_rnn', 'gan', and 'gan_for_defect_detection'
    main(product,data_dir, model_type=model_type, epochs=10, batch_size=32, sequence_length=5)
    model_type = 'cnn_rnn'  # Choose between 'cnn','dd', 'cnn_rnn', 'gan', and 'gan_for_defect_detection'
    main(product,data_dir, model_type=model_type, epochs=10, batch_size=32, sequence_length=5)
    model_type = 'gan'  # Choose between 'cnn','dd', 'cnn_rnn', 'gan', and 'gan_for_defect_detection'
    main(product,data_dir, model_type=model_type, epochs=10, batch_size=32, sequence_length=5)
    model_type = 'gan_for_defect_detection'  # Choose between 'cnn','dd', 'cnn_rnn', 'gan', and 'gan_for_defect_detection'
    main(product,data_dir, model_type=model_type, epochs=10, batch_size=32, sequence_length=5)
    #model_type = 'dd'  # Choose between 'cnn','dd', 'cnn_rnn', 'gan', and 'gan_for_defect_detection'
   # main(product,data_dir, model_type=model_type, epochs=10, batch_size=32, sequence_length=5)
```
