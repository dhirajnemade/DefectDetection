```python
import os
from PIL import Image
from collections import defaultdict
from prettytable import PrettyTable


def get_folder_names_in_directory(directory):
    # List to store the names of the folders
    folder_names = []

    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        # Add each directory name to the list
        folder_names.extend(dirs)
        
        # We only want the folders in the top level of the given directory
        break
    
    return folder_names

def create_image_dataset(source_dir, dest_dir, categories, subcategories, img_size=(128, 128)):
    """
    Creates an image dataset by organizing images into category and subcategory folders, resizing them,
    and providing a summary of the number of images processed.

    Parameters:
        source_dir (str): Path to the directory containing the images.
        dest_dir (str): Path to the destination directory where the dataset will be created.
        categories (list): List of category names (e.g., 'cats', 'dogs').
        subcategories (list): List of subcategory names (e.g., 'train', 'test', 'validation').
        img_size (tuple): Desired size for the images (width, height).

    Returns:
        None
    """
    # Dictionary to keep track of the count of processed images
    image_count = defaultdict(int)

    # Create the destination directory if it doesn't exist
    if not os.path.exists(dest_dir):
        os.makedirs(dest_dir)

    for category in categories:
        for subcategory in subcategories:
            # Create a folder for each category/subcategory in the destination directory
            subcategory_path = os.path.join(dest_dir, subcategory, category)
            if not os.path.exists(subcategory_path):
                os.makedirs(subcategory_path)

            # Path to the category/subcategory folder in the source directory
            source_category_path = os.path.join(source_dir, subcategory, category)

            # Check if the source category path exists
            if not os.path.exists(source_category_path):
                print(f"Source path '{source_category_path}' does not exist.")
                continue

            for img_name in os.listdir(source_category_path):
                img_path = os.path.join(source_category_path, img_name)

                try:
                    # Open and resize the image
                    with Image.open(img_path) as img:
                        img = img.resize(img_size)
                        img = img.convert("RGB")  # Convert to RGB if the image is not in RGB mode

                        # Save the resized image to the destination directory
                        img.save(os.path.join(subcategory_path, img_name))

                    # Increment the image count
                    image_count[(category, subcategory)] += 1

                    print(f"Processed {img_name} in {category}/{subcategory}")

                except Exception as e:
                    print(f"Could not process {img_name} in {category}/{subcategory}. Error: {e}")

    # Print the summary table
    print("\nSummary of Processed Images:\n")
    table = PrettyTable(["Category", "Subcategory", "Image Count"])
    for (category, subcategory), count in image_count.items():
        table.add_row([category, subcategory, count])
    
    print(table)

if __name__ == "__main__":
    # Example usage
    source_directory = "F:/MTVAD2/"
    destination_directory = "dataset/MTVAD/"
    subcategories_list = get_folder_names_in_directory(source_directory)  # Modify with your actual categories
    print(subcategories_list)
    categories_list = ["train", "test", "validation"]  # Subcategories like train, test, validation

    create_image_dataset(source_directory, destination_directory, categories_list, subcategories_list, img_size=(128, 128))
```

```python
from PIL import Image
import matplotlib.pyplot as plt

def display_original_and_resized_image(image_path, img_size=(128, 128)):
    """
    Displays the original and resized images side by side.

    Parameters:
        image_path (str): Path to the original image.
        img_size (tuple): Desired size for the resized image (width, height).

    Returns:
        None
    """
    # Open the original image
    original_image = Image.open(image_path)

    # Resize the image
    resized_image = original_image.resize(img_size)
    resized_image = resized_image.convert("RGB")  # Convert to RGB if necessary

    # Display the original and resized images side by side
    plt.figure(figsize=(10, 5))

    # Show original image
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(original_image)
    plt.axis('off')

    # Show resized image
    plt.subplot(1, 2, 2)
    plt.title(f"Resized Image ({img_size[0]}x{img_size[1]})")
    plt.imshow(resized_image)
    plt.axis('off')

    plt.show()

if __name__ == "__main__":
    # Example usage
    image_path = "dataset/MTVAD/wood/train/good_005.png"
    display_original_and_resized_image(image_path, img_size=(128, 128))
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to calculate and plot the 3D histogram for an image
def plot_3d_histogram(image_path, image_index):
    # Load the image using OpenCV
    image = cv2.imread(image_path)

    # Check if the image is loaded correctly
    if image is None:
        print(f"Error: Unable to load image {image_path}")
        return

    # Convert the image from BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Split the image into its RGB channels
    red_channel, green_channel, blue_channel = cv2.split(image_rgb)

    # Flatten the channels
    r = red_channel.ravel()
    g = green_channel.ravel()
    b = blue_channel.ravel()

    # Set the bins for the histogram
    hist_bins = 256
    hist_range = [0, 256]

    # Compute histograms for each channel
    hist_r, bins_r = np.histogram(r, bins=hist_bins, range=hist_range)
    hist_g, bins_g = np.histogram(g, bins=hist_bins, range=hist_range)
    hist_b, bins_b = np.histogram(b, bins=hist_bins, range=hist_range)

    # Create the x-axis values (intensity values)
    x_values = np.arange(hist_bins)

    # Create a figure for 3D plotting
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Plot the histograms in 3D space
    ax.bar(x_values, hist_r, zs=0, zdir='y', color='red', alpha=0.7, width=1.5, label=f"Red Channel")
    ax.bar(x_values, hist_g, zs=1, zdir='y', color='green', alpha=0.7, width=1.5, label=f"Green Channel")
    ax.bar(x_values, hist_b, zs=2, zdir='y', color='blue', alpha=0.7, width=1.5, label=f"Blue Channel")

    # Labels and Titles
    ax.set_xlabel('Pixel Intensity')
    ax.set_ylabel('Color Channel')
    ax.set_zlabel('Frequency')
    ax.set_yticks([0, 1, 2])
    ax.set_yticklabels(['Red', 'Green', 'Blue'])
    ax.set_title(f' Histogram of RGB Channels for Image {image_index + 1}')

    # Show the plot
    plt.show()

# List of image paths (replace these with your actual image paths)
image_paths = ['dataset/MTVAD/capsule/train/good_001.png', 'dataset/MTVAD/metal_nut/test/bent_006.png','dataset/MTVAD/leather/train/good_003.png', 'dataset/MTVAD/toothbrush/test/defective_007.png']

# Plot histograms for each image
for i, image_path in enumerate(image_paths):
    plot_3d_histogram(image_path, i)
```

```python
import cv2
import numpy as np
import os
import pandas as pd

# Function to perform statistical analysis on a single image
def analyze_image(image_path):
    # Load the image using OpenCV
    image = cv2.imread(image_path)

    if image is None:
        print(f"Error: Unable to load image {image_path}")
        return None

    # Convert the image to grayscale (optional, if you want to perform analysis on grayscale)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Calculate statistical properties
    mean = np.mean(gray_image)
    median = np.median(gray_image)
    std_dev = np.std(gray_image)
    variance = np.var(gray_image)
    min_value = np.min(gray_image)
    max_value = np.max(gray_image)

    # Return the statistics in a dictionary
    return {
        "mean": mean,
        "median": median,
        "std_dev": std_dev,
        "variance": variance,
        "min_value": min_value,
        "max_value": max_value
    }

# Function to analyze all images in a subcategory folder
def analyze_subcategory(subcategory_path):
    # List to store results for all images in the subcategory
    subcategory_stats = {
        "mean": [],
        "median": [],
        "std_dev": [],
        "variance": [],
        "min_value": [],
        "max_value": []
    }

    # Iterate over all image files in the subcategory folder
    for filename in os.listdir(subcategory_path):
        if filename.endswith(('.jpg', '.png', '.jpeg')):  # You can add more image formats if needed
            image_path = os.path.join(subcategory_path, filename)
            print(f"Analyzing {image_path}...")

            # Analyze the image and store the result
            image_stats = analyze_image(image_path)
            if image_stats:
                for key in subcategory_stats:
                    subcategory_stats[key].append(image_stats[key])

    # Calculate the mean of the statistics for the subcategory
    return {
        "mean": np.mean(subcategory_stats['mean']),
        "median": np.mean(subcategory_stats['median']),
        "std_dev": np.mean(subcategory_stats['std_dev']),
        "variance": np.mean(subcategory_stats['variance']),
        "min_value": np.mean(subcategory_stats['min_value']),
        "max_value": np.mean(subcategory_stats['max_value'])
    }

# Function to analyze all categories and subcategories
def analyze_all_categories(categories_root_path):
    # List to store statistics for each subcategory in each category
    all_stats = []

    # Iterate over each category folder
    for category in os.listdir(categories_root_path):
        category_path = os.path.join(categories_root_path, category)
        if os.path.isdir(category_path):  # Ensure it's a directory
            print(f"Analyzing category: {category}...")

            # Iterate over each subcategory in the category
            for subcategory in os.listdir(category_path):
                subcategory_path = os.path.join(category_path, subcategory)
                if os.path.isdir(subcategory_path):  # Ensure it's a directory
                    print(f"Analyzing subcategory: {subcategory} in category {category}...")

                    # Analyze the subcategory and get statistics
                    subcategory_stats = analyze_subcategory(subcategory_path)

                    # Append the results for this subcategory, along with the category and subcategory names
                    all_stats.append({
                        "category": category,
                        "subcategory": subcategory,
                        **subcategory_stats
                    })

    # Convert the dataset statistics to a DataFrame
    return pd.DataFrame(all_stats)

# Path to the root folder containing all categories (replace with the path to your root dataset folder)
categories_root_path = "dataset/MTVAD/"

# Analyze the dataset
dataset_statistics = analyze_all_categories(categories_root_path)

# Show the dataset statistics in a table
print(dataset_statistics)

# Optionally, save the dataset to a CSV file
dataset_statistics.to_csv('dataset_statistics.csv', index=False)
```
